```python
#Kody Chabaud
#Math 392: Numerical Methods
#Dr. Kuhn
#Program to apply the secant method to the following equation, taking in a tolerance of 10^-6
#f(x) = e^{-x} - 2x - ln(x) - x^{2}

import numpy as np
import math as m
#defining original function
def f(x):
  return m.e**(-x) -2*x - np.log(x) - x**2
  #return x**2 + 3 * m.e**(x) - 4 , tested with this polynomial, got correct root as well.
#defining secant method
def sec(p0, p1):
  tol = 10**(-6)
  maxit = 100
  for i in range(maxit):
    p2 = p1 - (f(p1)*(p1-p0))/(f(p1)-f(p0)) #secant formula
    if np.abs(p2 - p1) < tol:
      return p2
    else:
      p0 = p1
      p1 = p2
      sec(p0,p1)
p0 = .3 #Set our own initial guess
p1 = .4 #Set our own initial guess
p2 = sec(p0, p1)
print("The root of the following polynomial is: ", p2)
```
