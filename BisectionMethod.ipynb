```python
# This file solves f(x)=0 on [a,b] using the Bisection Method.
# Kody Chabaud 8/28/23
import math as m
a = 0
b = 1
n = 17 #num of iterations
for i in range(1, n+1):
  p = 1/2*(a+b)
  fa = m.exp(a) - a**2+3*a-2
  fp = m.exp(p) - p**2+3*p-2
  print(p)
  if fa * fp < 0:
    b = p
  else:
    a = p
```
