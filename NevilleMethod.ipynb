```python
#input: x,y, xr
#output: Q
import numpy as np


def neville(x,y,xr):
  n = len(x)
  Q = np.zeros([n,n])
  Q[:,0] = y #All rows, col 0
  for i in range(1,n):
    for j in range(1,i+1):
      Q[i,j] = ((xr-x[i-j])*Q[i,j-1] \
      - (xr-x[i])*Q[i-1,j-1]) \
      /(x[i]-x[i-j])
  return Q


x = np.array([8.1, 8.3, 8.6, 8.7])
y = np.array([17, 17.6, 18.5, 19])
xr = 8.4
Q = neville(x,y,xr)
print(Q)
```
