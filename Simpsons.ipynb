```python
#Kody Chabaud
#Numerical Methods
#Dr. Kuhn
#Due: 10/30/23
#Code that serves to implement Simpson's rule to approximate an area over a certain n value of evenly split sections.
import numpy as np

a = -1; b = 1
n = 80 #number of equal sections our area is split into
h = (b - a) / n
x = np.linspace(a,b,n+1)

f = np.exp(-(x**2)) #Our equation we're testing
#f = np.cos(x**2) #test equation
```

```python
def simpson(x, f, h):
  area = 0 #initialize to zero
  for i in range(1,n): # 1 to n-1
    area = (h/3) * (f[-1] + 4 * sum(f[1:n-1:2]) + 2 * sum(f[:n-2:2]) + f[1]) #Simpson's method
    return area
area = simpson(x,f,h)
print("Area =", area)#print statement
```
