```python
#Newtons Method for finding roots for f(x)=0
#Inputs: p0, f(x), f'(x)
#Outputs: root(p)
#Kody Chabaud
#9/6/23
#Math 392

#Import as needed
import numpy as np
#Define Functions
def f(x):
  return 2*x**3 + x**2 - x + 1
def fprime(x):
 return 6*x**2 + 2*x - 1

def newt(p):
  maxIt = 1000
  for i in range(maxIt):
    p_new = p - f(p)/fprime(p)
    if np.abs(p - p_new) == 0:
      print("Method Converged after", i, "Iterations")
      return p_new
    elif i == maxIt - 1:
      print("Method Diverged")
      return []
    #print(p_new)
    #Update
    p = p_new

#main
p0 = -1.2
p = newt(p0)
print(p)
```
