```python
import numpy as np
# Matrix form of Jacobi Iterative Method
A = np.array(([5, -2, 3], [-3, 9, 1], [2, -1, -7]))
b = np.array([-1, 2, 3])
x0 = np.array([0, 0, 0])
D = np.diag(A)
Dfull = np.identity(len(A)) * D
UplusL = -A + Dfull
Dinv = 1/D * np.identity(len(A))
T = np.matmul(Dinv, UplusL)
c = np.matmul(Dinv, b)
maxIt = 100
x = x0
for i in range(maxIt):

  xNew = np.matmul(T,x) + c
  # Error Check


  #Update
  x = xNew
print(x)
```
