```python
import numpy as np

def gauss(n):
  if n == 2:
    r = np.array([.5773502692, -.5773502692])
    c = np.array([1, 1])
  elif n == 3:
    r = np.array([.7745966692, 0, -.7745966692])
    c = np.array([.55555555555556, .88888888888889, .55555555555556])
  else:
    print("Gauss Rule not supported. Choose n = 2, ... , 6")
    return([], [])
  return([r,c])

n = 5;
[r,c] = gauss(n)
print(r)
```
