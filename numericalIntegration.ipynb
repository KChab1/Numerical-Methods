```python
import numpy as np
from scipy.integrate import trapz
a = 0; b = 1;
n = 71;
#Get nodal points
h = (b-a)/n
x = np.linspace(a, b, n+1)
#or this:
#x = np.arange(a, b+h, h)
f = np.cos(x**2)
#area = trapz(f,x)
#print(area)
```

```python
#coding trapz ourselves:

def trapezoidal(x,f, h):

  area = 0 #initialize to zero
  for i in range(1, n): #1 to n-1
    area = area + f[i]
  area = (h/2)*(f[0]+2*area+f[n])
  return area

area = trapezoidal(x,f,h)
print(area)
```
