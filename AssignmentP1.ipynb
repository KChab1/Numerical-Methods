```python
# This file solves f(x)=3(x-1)(x-1/2)(x-1) on [a,b] using the Bisection Method.
# Kody Chabaud
# 9/6/23
# Math 392: Assignment 1
# No use for math library so deleted.
a = -2 #left value of our interval
b = 1.5 #right value for our interval
n = 26 #num of iterations
for i in range(1, n+1): #iterates through 1,27 26 times total
  p = 1/2*(a+b)
  fa = 3*(a+1)*(a-1/2)*(a-1)
  fp = 3*(p+1)*(p-1/2)*(p-1)
  print(p)
  if fa * fp < 0:
    b = p
  else:
    a = p
```
