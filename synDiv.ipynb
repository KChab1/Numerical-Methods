```python
#Inputs: P (Array), x0 (Guess)
#Outputs: Q (Array), remainder

import numpy as np
def synDiv(P, x):
  n = len(P)
  Q = np.zeros(n-1)
  Q[0] = P[0]
  for i in range(1,n):
    if i == (n-1):
      rem = Q[i-1]*x + P[i]
      return[Q,rem]
    else:
      Q[i] = Q[i-1]*x + P[i]

P = np.array([-1, 0, 0, 3, 0,0, -8, 9])
x0 = 1.5 #guess
#Get P at x0
[Q, rem] = synDiv(P,x0)
[Q2, rem2] = synDiv(Q,x0)
print('P at -2:', rem)
print('P prime at -2:', rem2)
```
