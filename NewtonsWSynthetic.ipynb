```python
#Newton's Method with Synthetic Division
#Kody Chabaud
#Dr. Kuhn
#392 Assignment 3
#Due 10/4/23

import numpy as np

#equation: x**6 + 5x**3 - 85x - 136

def synDiv(P, x):
  n = len(P)
  Q = np.zeros(n-1)
  Q[0] = P[0]
  for i in range(1,n):
    if i == (n-1):
      rem = Q[i-1]*x + P[i]
    else:
      Q[i] = Q[i-1]*x + P[i]
  return[Q,rem]

def newt(p, eq):

  maxIt = 1000

  print("Initial Guess", p)

  for i in range(maxIt):
    [Q, rem] = synDiv(eq, p)
    [_, rem2] = synDiv(Q, p)
    pnew = p - rem/rem2
    print("New PNew: ", pnew)
    if np.abs(p - pnew) <= tol:
      print("Method Converged After ", i , "Iterations: ", pnew)
      return pnew
    elif i == maxIt - 1:
      print("Method Diverged")
      return []
    #print(p_new)
    #Update
    p = pnew

P = np.array([1, 0, 0, 5, 0, -85, -136]) #feed your coefficients from equation into this array
p0 = 4 # initial guess
tol = 10**-5 #tolerance for accuracy
newt(p0, P)
```
